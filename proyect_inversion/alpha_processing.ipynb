```python
# Seleccionar las columnas relevantes
df = df_MSFT.select(
    col("Meta Data.*"),
    from_json(col("Time Series (Daily)"), "map<string, struct<1. open: string, 2. high: string, 3. low: string, 4. close: string, 5. volume: string>>").alias("Time_Series")
)

# Expandir las columnas de la estructura Time Series
for i in range(1, 6):
    df_MSFT = df_MSFT.withColumn(f"open_{i}", col("Time_Series").getItem(i + ". open"))
    df_MSFT = df_MSFT.withColumn(f"high_{i}", col("Time_Series").getItem(i + ". high"))
    df_MSFT = df_MSFT.withColumn(f"low_{i}", col("Time_Series").getItem(i + ". low"))
    df_MSFT = df_MSFT.withColumn(f"close_{i}", col("Time_Series").getItem(i + ". close"))
    df_MSFT = df_MSFT.withColumn(f"volume_{i}", col("Time_Series").getItem(i + ". volume"))

# Eliminar columnas innecesarias
df_MSFT = df_MSFT.drop("Time_Series")

# Mostrar el DataFrame
df_MSFT.show(truncate=False)
```

```python
# Obtener la lista de archivos JSON en la carpeta
path = r"C:\Users\CARLES\Documents\proyect_inversion\symbols" 
archivos_json = [archivo for archivo in os.listdir(path) if archivo.endswith('.json')]

# Inicializar un DataFrame vaco
df = pd.DataFrame()
```

```python
dfs = []

# Iterar sobre los archivos JSON y cargarlos en DataFrames individuales
for archivo in archivos_json:
    ruta_archivo = os.path.join(path, archivo)
    datos_json = pd.read_json(ruta_archivo)
    dfs.append(datos_json)

# Concatenar todos los DataFrames en uno solo
df = pd.concat(dfs, ignore_index=True)

df
```
