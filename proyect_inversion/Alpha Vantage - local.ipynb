```python
# INCREMENTAL FUNCTIONS
# Función para obtener la última fecha de actualización de un archivo JSON
def get_last_refreshed_date(file_path, symbol):
    
    file_path_symbol = f"{file_path}\\symbols\\{symbol}.json"

    with open(file_path_symbol, 'r') as file:
        data = json.load(file)

    return data['Meta Data']['3. Last Refreshed']

# Función para guardar la última fecha de actualización en un archivo JSON
def save_last_refreshed_date(file_path, date):
    with open(file_path, 'r') as file:
        data = json.load(file)
    data['Meta Data']['3. Last Refreshed'] = date
    with open(file_path, 'w') as file:
        json.dump(data, file)
```

```python
# SYMBOLS PARAMETERS
interval_symbol = 'daily'
outputsize_symbol = 'full'
function_symbol = 'TIME_SERIES_DAILY'

for i in symbols:
    path = f"{local}\\symbols\\{i}.json"
    url = f"https://www.alphavantage.co/query?function={function_symbol}&symbol={i}&outputsize={outputsize_symbol}&apikey={api_key}"
    response = requests.get(url)
    data = response.json()

    if response.status_code == 200:
        # Verifica si el archivo ya existe y ciérralo si está abierto
        if os.path.exists(path):
            os.remove(path)

        with open(path, 'w') as file:
            json.dump(data, file)

        print("Los datos para", i, "se han guardado exitosamente en", path)
    else:
        print("Error al obtener los datos para", i, ". Código de estado:", response.status_code)
```

```python
def obtener_historical_quotes(symbol, data):
    data_m['Meta Data']['2. Symbol'] == symbol
    historical_quotes = []

    for date, values in data["Time Series (Daily)"].items():
        # Verificar si la fecha es válida antes de crear un objeto Quote
        if date is not None:
            # Convertir la fecha de cadena a datetime
            date_datetime = datetime.strptime(date, "%Y-%m-%d")
            
            # Crear el objeto Quote
            quote = Quote(symbol=symbol, date=date_datetime, open=float(values["1. open"]), high=float(values["2. high"]),
                          low=float(values["3. low"]), close=float(values["4. close"]), volume=int(values["5. volume"]))
            
            # Agregar el objeto Quote a la lista de citas históricas
            historical_quotes.append(quote)
    
    return historical_quotes
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
import pandas_ta as ta
import matplotlib.dates as mdates
from datetime import datetime, timedelta

# Define the stock symbol and timeframe
symbol = 'DIS'
end_date = datetime.today()
start_date = end_date - timedelta(days=120)  # 4 months before today

# Fetch stock data using yfinance
stock_data = yf.download(symbol, start=start_date, end=end_date)

# Calculate technical indicators using pandas-ta
stock_data.ta.macd(append=True)
stock_data.ta.rsi(append=True)
stock_data.ta.bbands(append=True)
stock_data.ta.obv(append=True)

# Calculate additional technical indicators
stock_data.ta.sma(length=20, append=True)
stock_data.ta.ema(length=50, append=True)
stock_data.ta.stoch(append=True)
stock_data.ta.adx(append=True)

# Calculate other indicators
stock_data.ta.willr(append=True)
stock_data.ta.cmf(append=True)
stock_data.ta.psar(append=True)

#convert OBV to million
stock_data['OBV_in_million'] =  stock_data['OBV']/1e7
stock_data['MACD_histogram_12_26_9'] =  stock_data['MACDh_12_26_9'] # not to confuse chatGTP

# Summarize technical indicators for the last day
last_day_summary = stock_data.iloc[-1][['Adj Close',
    'MACD_12_26_9','MACD_histogram_12_26_9', 'RSI_14', 'BBL_5_2.0', 'BBM_5_2.0', 'BBU_5_2.0','SMA_20', 'EMA_50','OBV_in_million', 'STOCHk_14_3_3', 
    'STOCHd_14_3_3', 'ADX_14',  'WILLR_14', 'CMF_20', 
    'PSARl_0.02_0.2', 'PSARs_0.02_0.2'
]]

print("Summary of Technical Indicators for the Last Day:")
print(last_day_summary)
```
